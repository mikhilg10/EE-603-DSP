```
#problem 2(a)plot function $/delta$[n]
import numpy as np
import pylab

n=np.arange(-8,9)
l=np.size(n)
pylab.close('all')
imp=np.zeros(l)
ind=np.where(n==0)
imp[ind]=1
pylab.stem(n,imp)
pylab.xlabel('x samples')
pylab.ylabel('Amplitude')
pylab.grid()
pylab.show()
```

```
#problem 2(f)plot function $/delta$[n]
import numpy as np
import pylab

n=np.arange(-8,9)
l=np.size(n)
function=np.ones(l)
sinc=np.sin(np.pi/2*n)/(np.pi*n)
prob=function*sinc
lim=8
prob[lim]=1 #taking limit for n=0
pylab.stem(n,prob)
pylab.xlabel('x samples')
pylab.ylabel('Amplitude')
pylab.grid()
pylab.show()
```
