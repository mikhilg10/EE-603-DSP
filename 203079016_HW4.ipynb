```
#Question 5
from scipy.signal import kaiserord,firwin, freqz,remez
import matplotlib.pyplot as plt
import numpy as np
import remezord
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

numtaps, beta = kaiserord(75,0.05)
rp = 1;
rs=75 
f = np.array([650,700]) 
fs = 2000; 
min = (10**(rp/20) - 1)
max = (10**(rp/20) + 1)
a=[1,0]
dev = np.array([min/max, 10**(-rs/20)]) 
[n,fo,ao,w] = remezord.remezord(f,a,dev,fs)     
lpf = signal.remez( n,fo,ao,w)
w, h = signal.freqz(lpf, [1], worN=2000)
plt.plot(w, 20 * np.log10(abs(h)))
plt.grid(True)
print("5b(ii) kaiser window order is", numtaps)
N, Wn = signal.ellipord(0.65, 0.7, 1, 75, False)
print('5b(iii) ORDER OF FILTER:',N)
print( '5bi remezord ORDER OF FILTER:',n)
```

```
#5bi
numtaps, beta = kaiserord(75,0.05)
print("order is", numtaps/2)
cutoff=0.675
taps = firwin(numtaps, cutoff, window=('kaiser', beta),
              scale=False)
w, h = freqz(taps, worN=4192)
plt.figure(1)
xticks =np.linspace(-np.pi, np.pi,num=10)
xtickslabel = ['$-\pi$', '$-3\pi/4$', '$-\pi/2$','$-\pi/4$', '$0$', '$\pi/4$', '$\pi/2$','$3\pi/4$','$\pi$']
plt.title( 'Magnitude kaiser LPF')
plt.plot(w, 20 * np.log10(abs(h)))
plt.xticks(xticks, xtickslabel)
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.figure(2)
plt.plot(w,  (np.angle(h)))
plt.xticks(xticks, xtickslabel)
plt.title( 'Phase response kaiser LPF')
plt.ylabel('PhaseAngle')
plt.grid(which='both', axis='both')
print('coefficients of z transform')
for i in range(len(taps)):
    print(taps[i])
#denominator is 1
```

```
#5bii
dev = np.array([min/max, 10**(-rs/20)]) 
[n,fo,ao,w] = remezord.remezord(f,a,dev,fs)     
lpf = remez( n,fo,ao,w)
w, h = freqz(lpf, [1], worN=2000)
taps = firwin(n,0.675)
print('coefficients of z transform')
for i in range(len(taps)):
    print(taps[i])
plt.figure(1)
plt.plot(w, 20 * np.log10(abs(h)))
plt.xticks(xticks, xtickslabel)
plt.title('Mag response of FIR design using  Parks-McClellan algorithm')
plt.ylabel('Abs mag (db)')
plt.grid(which='both', axis='both')
plt.figure(2)
plt.plot (w, ( (np.angle(h))))
plt.xticks(xticks, xtickslabel)
plt.title('Phase response of FIR design using  Parks-McClellan algorithm')
plt.ylabel('PHASE')
plt.grid(which='both', axis='both')
```

```
#6a
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

fs = 22050.0       # Sample rate, Hz
cutoff = 4000.0    # Desired cutoff frequency, Hz
trans_width = 200  # Width of transition from pass band to stop band, Hz
numtaps = 200      # Size of the FIR filter.
lpf = signal.remez(numtaps, [0, cutoff, cutoff + trans_width, 0.5*fs], [1, 0], Hz=fs)
w, h = signal.freqz(lpf, [1], worN=2000)
plt.plot(0.5*fs*w/np.pi,np.abs(h))
plt.xlabel('Freq(Hz)')
plt.ylabel('abs magn value')
plt.title('LPF 4K')
plt.grid(True)
```

```
#6a
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
fs =44100        # Sample rate, Hz
band = [4000, 8000]  # Desired pass band, Hz
trans_width = 200    # Width of transition from pass band to stop band, Hz
numtaps = 10        # Size of the FIR filter.
edges = [0, band[0] - trans_width, band[0], band[1],
         band[1] + trans_width, 0.5*fs]
bpf4k = signal.remez(numtaps, edges, [0, 1, 0], Hz=fs)
w, h = signal.freqz(bpf4k, [1], worN=4192)
plt.plot(0.5*fs*w/np.pi,np.abs(h))
plt.xlabel('Freq (Hz)')
plt.ylabel('abs magn value')
plt.title('BPF 4K-8K')
plt.grid(True)
```
